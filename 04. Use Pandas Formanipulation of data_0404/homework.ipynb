```python
customerData["PRODUCT_AGE_NEW"] = np.where(customerData["PRODUCTAGE"] < 1 , 1,\
                                  np.where(customerData["PRODUCTAGE"] < 2,2,\
                                  np.where(customerData["PRODUCTAGE"] < 3,3,5)))
customerData.head(3)
```

```python
kcsn["QTY_NEW"] = np.where(kcsn["QTY"] < 0,0,kcsn["QTY"])
kcsn["QTY_NEW"].min()
```

```python
kcsn["YEAR"] = kcsn["YEARWEEK"].str[:4]
kcsn["WEEK"] = kcsn["YEARWEEK"].str[4:]
refinedSelloutData = kcsn[kcsn["WEEK"].astype(int) <=52]
refinedSelloutData["WEEK"].max()
```

```python
EMImax = customerData.groupby(["EMI"]).max()
EMImax
```

```python
sortedData = refinedSelloutData.sort_values(["REGIONID", "PRODUCT", "YEARWEEK"])
sortedData.head()
```

```python
groupData = sortedData.groupby(["REGIONID", "PRODUCT","YEAR"]).agg(["mean", "std"])["QTY_NEW"]
groupData.columns = ["QTY_MEAN", "QTY_STD"]
```
